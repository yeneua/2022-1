```python
# pivot table
import pandas as pd
df = pd.DataFrame([ ['20180901', 'A', 10],
                   ['20180901', 'B', 100],
                   ['20180901','C', 1000],
                   ['20180902', 'A', 20],
                   ['20180902', 'B', 200],
                   ['20180902', 'C', 2000],
                   ['20180903', 'A', 30],
                   ['20180903', 'B', 300],
                   ['20180903', 'C', 3000],
], columns=['date', 'typecode', 'volume'])
```

```python
df
```

```python
df_pivoted1 = df.pivot(index = 'date', columns = 'typecode', values = 'volume')
```
