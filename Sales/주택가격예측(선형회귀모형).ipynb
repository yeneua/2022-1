```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston #sklearn패키지에 내장된 데이터
boston = load_boston()
```

데이터 수집, 준비 및 탐색 - 2. 데이터셋 구성 확인

```python
print(boston.DESCR)
# 데이터셋 설명
# 설명변수만 나와있음
# target 변수 - median value
```

```python
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
boston_df.head()
```

```python
boston_df.shape # 변수 13개. 타깃변수는 따로 저장되어있음
```

```python
boston_df['PRICE'] = boston.target #타겟변수 불러오기
boston_df.head()
```

```python
boston_df.shape #타겟변수 추가됨 - 14개 변수
```

```python
boston_df.info()
```

분석 모델 구축, 결과 분석 및 시각화 - 1. 선형회귀를 이용해 분석 모델 구축하기
- 사이킷런의 선형 분석 모델 패키지에서 선형 회귀를 이용하여 분석 모델을 구축

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# x, y 분할
Y = boston_df['PRICE']
X = boston_df.drop(['PRICE'], axis = 1, inplace = False)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 156)
```

```python
# 모델 생성
lr = LinearRegression()
```

```python
# 모델 훈련(선형 회귀 분석)
lr.fit(X_train, Y_train)
```

```python
# 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

```python
Y_predict #array 형태로
```

분석 모델 구축, 결과 분석 및 시각화 - 1. 선형 회귀를 이용해 분석 모델 구축하기
- 선형 회귀 분석 모델을 평가 지표를 통해 평가하고 회귀 계수를 확인하여 피처의 영향을 분석

```python
mse = mean_squared_error(Y_test,Y_predict)
rmse = np.sqrt(mse)
r = r2_score(Y_test, Y_predict)
print("mse : %f, rmse : %f, r^2(variance score) : %f" %(mse, rmse, r))
```

```python
print('Y절편 값: ',lr.intercept_)
print('회귀 계수 값:', np.round(lr.coef_, 1))
```

```python
# 회귀 계수 정렬
coef = pd.Series(data = np.round(lr.coef_, 1), index = X.columns)
coef.sort_values(ascending = False)
```

회귀 분석 결과를 산점도 + 선형 회귀 그래프로 시각화하기 - 2. 선형회귀를 이용해 분석 모델 구축하기

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axs = plt.subplots(figsize = (16, 16), ncols = 3, nrows = 5)

x_features = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']

for i, feature in enumerate(x_features):
  row = int(i/3)
  col = i%3
  sns.regplot(x = feature, y = 'PRICE', data = boston_df, ax = axs[row][col])
```
