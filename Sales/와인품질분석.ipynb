데이터 준비 - 1.csv 파일 다운로드

```python
import pandas as pd
red_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep = ';', header=0, engine='python')
white_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep=';',header=0,engine='python')
```

데이터 모델링 - 2. t-검정과 회귀 분석으로 그룹 비교하기

```python
from scipy import stats #t검정
from statsmodels.formula.api import ols, glm #회귀분석
```

```python
# t-test
red_wine_quality = wine.loc[wine['type']=='red','quality']
white_wine_quality=wine.loc[wine['type']=='white','quality']
stats.ttest_ind(red_wine_quality,white_wine_quality,equal_var=False)
```

```python
#회귀분석
Rformula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar+chlorides+free_sulfur_dioxide+total_sulfur_dioxide+density+pH+sulphates+alcohol'
regression_result=ols(Rformula, data=wine).fit()
regression_result.summary()
```

데이터 모델링 - 3. 회귀분석 모델로 새로운 샘플의 품질 등급 예측하기

```python
sample1 = wine[wine.columns.difference(['quality','type'])] #quality,type빼고 나머지를 저장
sample1 = sample1[0:5][:] #5개데이터
```

```python
sample1_predict = regression_result.predict(sample1) #regression_result: 아까만든 회귀모델. predict로 예측
sample1_predict #예측값
```

z검정, t검정

```python
red_wine_quality = wine.loc[wine['type'] == 'red','quality']
white_wine_quality = wine.loc[wine['type'] == 'white','quality']
```

```python
from statsmodels.stats.weightstats import ztest as ztest
ztest(red_wine_quality,white_wine_quality,value=0) #z검정 => 결과로 p값이 나온다
```

```python
stats.ttest_ind(red_wine_quality,white_wine_quality,equal_var=False) #t검정, 이분산 가정
```

```python
stats.ttest_ind(red_wine_quality,white_wine_quality,equal_var=True) #t검정, 등분산 가정
```
