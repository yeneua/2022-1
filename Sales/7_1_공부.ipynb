```python
import pandas as pd
import numpy as np 
url=("https://raw.github.com/pandas-dev/pandas/main/pandas/tests/io/data/csv/tips.csv")
tips = pd.read_csv(url)
tips.head() ## 데이터프레임의 앞에 5개만 보여주는 것
```

변수추가 <br>
tip을 total_bill로 나눈 tip_rate 변수 추가

```python
#방법1 - 데이터명.assign(변수명=계산식 및 내용)
# *assign으로 하면 실제 데이터에 저장X. 추가된 데이터 얻으려면 기존 변수에 다시 넣어주기
tips.assign(tip_rate = tips['tip']/tips['total_bill']).head()
```

```python
#방법2 - 데이터명['변수명'] = 계산식 및 내용
tips['tip_rate1'] = tips['tip']/tips['total_bill']
```

```python
#방법3 - 데이터명.insert(인덱스위치, '변수명', 계산식 및 내용)
tips.insert(8,'tip_rate2',tips['tip']/tips['total_bill'])
```

group by<br> count(), size()

```python
tips.groupby('sex').size() #성별별로 count
```

```python
tips.groupby('smoker').size() #size()는 sql에서 group by로 count를 한 것과 같다.
```

```python
tips.groupby('sex').count() #성별별로
```

```python
tips.groupby('smoker').count  #smoker별로
```

```python
tips.groupby('sex')['total_bill'].count() #성별로 grouping한 것에서 total_bill에 대해 count()한 결과 
```

```python
#count() applies the function to each column, returning the number of not null records
# count()는 null값을 포함하지 않고 계산, size()는 null값을 포함하여 계산
# => size()는 호출된 객체의 크기 또는 길이 반환
```

여러함수를 적용하여 groupby 집계
- day별로 tip의 평균과 행의 개수 count
- smoker,day별로 tip의 개수와 평균

```python
#day를 groupby 기준으로 정하고, tip에는 mean() 적용, day에는 size()적용
tips.groupby('day').agg({'tip':np.mean, 'day':np.size})
```

```python
tips.groupby('day').agg({'tip':'mean','day':'size'})
```

```python
tips.groupby('day').agg({('tip','mean'),('day','size')}) #numpy를 사용하지 않으면 모든 열에 대하여 함수적용되어나옴 ?? 사실 뭔지 몰겟다
```

```python
tips.groupby(['smoker','day']).agg({'tip':[np.size,np.mean]})
```

데이터프레임 조인

```python
df1=pd.DataFrame({'key' : ['A','B','C','D'],'value':np.random.rand(4)})
df2=pd.DataFrame({'key' : ['B','D','D','E'],'value':np.random.rand(4)})
```

```python
df1
```

```python
df2
```

```python
#merge  - default:innerjoin
pd.merge(df1,df2,on='key')
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,on='key',how='right')
```
