###코딩변경
기존의 데이터 값을 새로운 기준으로 

1,2,3,4,5,6,7,8 학력 1,2:고졸. 3,4:대졸, 5,6,7,8:석박<br>
1     2     3      => 3개의 집단으로 구분(범주화)

60,80,101,124,234, ... <br>
100미만 = 1<br>
101 ~ 200 =2 <br>
301 ~  = 3<br>
=> 어떤 기준에 따라서 연속형 자료 범주화

```python
1,2,3,4,5,6,7,8 #학력 1,2:고졸. 3,4:대졸, 5,6,7,8:석박
1   2   3      #=> 3개의 집단으로 구분(범주화)
```

```python
#파이썬에서 코딩변경 하려면 함수정의(definition)필요
```

```python
def edu_group(series):
  if series == 2 :
    return 1
  elif series == 3 or series==4:
    return 2
  elif series >=5:
    return 3
```

```python
df['eduM'] = df['edu'].apply(edu_group) #df데이터셋에 추가
```

```python
df.head()
```

```python
df['eduM'].value_counts(sort=False) #새로운변수 빈도 확인  #왜 sort 안되냐고 ..
```
